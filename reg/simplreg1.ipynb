```python
import pandas as pd
import numpy as np

data = pd.read_csv('Experience-Salary.csv')

data.head()
```

```python
data
```

```python
x = data[["exp(in months)"]].copy()
y = data[["salary(in thousands)"]].copy()
```

```python
x["int"] = 1

x = x[["int", "exp(in months)"]]

x
```

```python
coeff = np.linalg.inv(x.T @ x) @ x.T @ y

coeff
```

```python
coeff.index = x.columns
coeff
```

```python
pred = x @ coeff
pred
```

```python
ssr = ((y - pred) ** 2).sum()
sst = ((y - y.mean()) ** 2).sum()

acc = (1 - (ssr/sst)) * 100
```

```python
acc
```
