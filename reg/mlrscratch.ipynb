```python
import pandas as pd

dt = pd.read_csv("Student_Performance.csv")
```

```python
dt
```

```python
print(dt.columns)
```

```python
dt['Extracurricular Activities'] = dt['Extracurricular Activities'].map({"Yes":1, "No":0})
dt.head()
```

```python
x = dt[['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']].copy()
y = dt[['Performance Index']].copy()

x['int'] = 1

x = x[['int', 'Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']]

x.head()
```

```python
import numpy as np
coeff = np.linalg.inv(x.T @ x) @ x.T @ y
coeff
```

```python
coeff.index = x.columns
coeff
```

```python
pred = x @ coeff
pred
```

```python
ssr = ((y - pred) ** 2).sum()
sst = ((y - y.mean()) ** 2).sum()

acc = (1 - (ssr/sst)) * 100

acc
```
